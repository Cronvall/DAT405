```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
import random
```

```python
data = pd.read_csv('data.csv')
data.head()
```

## 2A

```python
def cluster(n, df):
    # creates n number of clusters and assign each row a cluster id
    km = KMeans(n_clusters=n)
    y_predicted = km.fit_predict(df[['phi','psi']])
    df['cluster'] = y_predicted
    
    # Draws each cluster by reading all rows but only drawing rows with correct cluster id
    for i in range(n):
        cluster = df[df.cluster==i]
        col= hexadecimal = ["#"+''.join([random.choice('ABCDEF0123456789') for i in range(6)])]
        plt.scatter(cluster['phi'],cluster['psi'],color=col, s=1)
        
    # Draws cluster centroids and a legend to help understand what the stars are
    plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
    plt.legend()
    
    # Plot labels and title
    plt.title(f"Number of clusters: {n}")
    plt.xlabel('phi angle (degrees)')
    plt.ylabel('psi angle (degress)')
```

```python
cluster(3, data)
```

```python
cluster(4, data)
```

We determined that the appropiate number of clusters is three. This since that is the number of clusters we ourself would determine their to be.
Our second argument is that if we compare our plot to the Ramachandran plot we can see that the clusters of: 
 - Alpha helicals
 - Beta strands
 - Turns
 
The other number of clusters we were debating is four clusters. This since their is a large collection of dots in the bottom left. But if we once again look at the Ramachandran plot we can see that these dots belong with the cluster of beta strands. Why the K-means algorithm doesn't include these dots in the same cluster is because of that it doesn't take into account the fact that we are measuring angles and that they are circular.

## 2D

```python
changed_df = data
for i,row in changed_df.iterrows():
    if row.psi <= -100:
        changed_df.at[i,'psi'] = (180 + row.psi)
```

```python
changed_df = changed_df.sample(frac=0.5)
cluster(3,changed_df)
```

The code above is moving all dots with a psi angle below or equal to -100 degree to the top of the plot. </br>
By doing this we ease the understanding of what we ment that the bottom left dots actually belonged in the beta strands cluster.
One thing that we did not expect happend is the change of placement of the centroid for the turns cluster. We think this happens since we move all of the dots with a psi angle bellow -100 degrees no matter it's phi angle. </br>
So in short we have displaced the psi angles to ease the understanding of why we have three clusters and how the dots in each cluster is related, especially in the beta strands cluster.
