```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
phi = data['phi']
psi = data['psi']
```

```python
plt.scatter(phi,psi, s=1)
plt.plot()
```

```python
roof = int(round(max(psi.max(),phi.max()),0))
array = np.empty([roof*2+1,roof*2+1])

for i in range(len(phi)):
    x = round(phi[i], 0)
    y = round(psi[i], 0)
    if x >= 0: x = x + x
    else: x = abs(x)
    if y >= 0: y = y + y
    else: y = abs(y)

    array[int(x)][int(y)] = array[int(x)][int(y)] + 1

```

```python
plt.imshow(array, cmap='hot', interpolation='nearest')
plt.gca().invert_yaxis()
plt.show()
```
