```python
#Imports needed during the problems
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
#Problem 2
# Read the three csv files with data for corruption, happiness and trust.
corruption_data = pd.read_csv("./data/corruption.csv")
happiness_data = pd.read_csv("./data/happiness.csv")
trust_data = pd.read_csv("./data/trust.csv")
```

```python
#Corruption data set
#‘Corruption Perception Index’, which is a indicator of corruption of the entire world.
#The Corruption Perception Index scores countries on a scale of 0-100, where 0 means that 
#a country is perceived as highly corrupt and 100 means that a country is perceived as very clean. 

#Spain from 2012 to 2018
xValues_spain = corruption_data['Year'][1024:1031]
yValues_spain = corruption_data['Corruption Perception Index - Transparency International (2018)'][1024:1031]
plt.scatter(xValues_spain, yValues_spain, c='b')
plt.title("Spain")
plt.show()

#Sweden from 2012 to 2018
xValues_sweden = corruption_data['Year'][1052:1059]
yValues_sweden = corruption_data['Corruption Perception Index - Transparency International (2018)'][1052:1059]
plt.scatter(xValues_sweden, yValues_sweden, c='r')
plt.title("Sweden")
plt.show()

#Spain vs Sweden from 2012 to 2018
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(xValues_spain, yValues_spain, s=10, c='b', marker="s", label='Spain')
ax1.scatter(xValues_sweden,yValues_sweden, s=10, c='r', marker="o", label='Sweden')
plt.legend(loc='center')
plt.show()

```

```python
#The World Happiness Report is a source of cross-country data and research on 
#self-reported life satisfaction. It includes the ‘happiness scores’ per country.
#The life evaluation is numbered from 0 at the bottom to 10 at the top. The top 
#represents the best possible life for you and the bottom of the ladder represents 
#the worst possible life. 

russia = happiness_data['Life satisfaction in Cantril Ladder (World Happiness Report 2022)'][1668:1685]
usa = happiness_data['Life satisfaction in Cantril Ladder (World Happiness Report 2022)'][2115:2132]

#U.S.A from 2004 to 2020
plt.boxplot(usa)
plt.xticks([1], ['U.S.A'])
plt.title("U.S.A")
plt.show()

#Russia from 2004 to 2020
plt.boxplot(russia)
plt.xticks([1], ['Russia'])
plt.title("Russia")
plt.show()

#U.S.A vs Russia from 2004 to 2020
plt.boxplot([russia,usa])
plt.xticks([1, 2], ['Russia', 'U.S.A'])
plt.title("Russia vs U.S.A")
plt.show()
```

```python
#This survey allows cross-country comparisons of self-reported trust attitudes. 
#This share different countries and in different years, the evaluation is numbered 
#from 0 at the bottom to 100 at the top. The top represents the best possible trust 
#attitudes and the bottom of the represents the worst possible.

#Mexico vs Switzerland vs China
mexico_data = trust_data[114:120]
switzerland_data = trust_data[198:201]
china_data = trust_data[40:45]

plt.plot(mexico_data['Year'], mexico_data['Trust in others (World Values Survey (2014))'], label = 'Mexico') 
plt.plot(switzerland_data['Year'], switzerland_data['Trust in others (World Values Survey (2014))'], label = 'Switzerland') 
plt.plot(china_data['Year'], china_data['Trust in others (World Values Survey (2014))'], label = 'China') 
plt.legend(loc=2, prop={'size': 6})
plt.show()
```
